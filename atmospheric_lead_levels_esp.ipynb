# Evolución de los niveles de plomo (Pb) en la atmósfera terrestre

## Introducción

El [plomo (Pb)](https://es.wikipedia.org/wiki/Plomo) es el elemento de la tabla periódica con número atómico 82. Se trata de un metal pesado de gran maleabilidad y punto de fundición bajo. Por sus características, ha sido y es de gran utilidad industrial. Se sabe, sin embargo, que es tóxico y que puede causar anemía, perturbaciones en el sistema nervioso, disminución de las habilidades de aprendizaje e incluso aumento en los niveles de agresividad.

![](./media/1134px-Lead_electrolytic_and_1cm3_cube.jpg)

La producción de Pb en la era común tubo un gran pico a finales de la época clásica, con el esplendor del Imperio Romano. Sin embargo, este pico no se compara al logrado con la revolución industrial y con el posterior uso de un tipo de gasolina de alto octanaje que utilizaba plomo como aditivo antidetonante.

En la era común, el único periodo en que se han logrado niveles de Pb atmosférico similares a los naturales fue durante la [peste negra](https://es.wikipedia.org/wiki/Peste_negra): la mayor pandemía de la historia de Europa. 
  
En el presente análisis se hará uso de un registro de deposición de Pb armosférico para observar la evolución de los niveles de este metal en la atmósfera terrestre.

El dataset utilizado será dividido en secciones que corresponden a distintos peridos históricos y cada sección será analizada por separado. Los picos más notorios en la cantidad de Pb en la atmósfera serán señalados mediante anotaciones. 

Finalmente, será creada y presentada una visualización que abarque todos los datos disponibles en el dataset, con la intención de que muestre claramente las conclusiones más importantes de éste análisis y que pueda observarse, en una sola imágen, la evolución de los niveles de Pb en los ~$2000$ años de la era común.

## Acerca de los datos

Los datos aquí utilizados pertencen al registro de deposición de Pb atmosférico realizado para el estudio *[Next-generation ice core technology reveals true minimum natural levels of lead (Pb) in the atmosphere: Insights from the Black Death](https://agupubs.onlinelibrary.wiley.com/doi/10.1002/2017GH000064)*, publicado por la American Geophysical Union (AGU) en 2017. En concreto será utilizado el archivo *gh227-sup-0004-Dataset S3 Discrete_ICP-MS_Pb.xls*, brindado como información de soporte para el estudio en su versión online. Este archivo contiene un registro de deposición de Pb obtenido de un núcleo de hielo de ~$72 \text{ m}$, extraído del glaciar Colle Gnifetti, el cual se encuentra a ~$4450 \text{ m}$ sobre el nivel del mar, en los Alpes suizo-italianos.

Las observaciones fueron obtenidas, según se explica en el artículo del estudio, mediante espectrometría de masas con plasma acoplado inductivamente (ICP-MS), de alta resolución. Estas observaciones abarcan los ~$2000$ años de la era común.

## Evolución del nivel de Pb en la atmósfera

### Importar los datos

Durante la época clásica el mayor productor de Pb llegó a ser el Imperio Romano. Este metal se obtenía principalmente como un subproducto de la producción de plata, la cual se extraía de la [galena](https://es.wikipedia.org/wiki/Galena).

## Edad media

Tras la caida del Imperio Romano de Occidente la producción de Pb disminuyó en Europa y no fue hasta  los siglos XI y XII cuando volvió a utilizarse, principalmente en pintura, en tuberías y como recubrimento de edificios.

## Revolución industrial

Durante la revolución industrial el uso de Pb en la industria aumentó como resultado de su demanda para la creación de tuberías y su uso como aditivo de pinturas de alta resistencia al agua. Esto causó, por supuesto, un gran aumento en las personas que estaban expuestas a este metal y que padecierón de intoxicación. Los primeros estudios modernos acerca de la [plumbosis](https://es.wikipedia.org/wiki/Saturnismo) (intoxicación por plomo) y las primeras regulaciónes legales para mejorar las condiciones de los trabajadores que debían estar expuestos Pb se produjeron a finales del siglo XIX. 

## Uso de gasolina con TEL

Para comienzos del siglo XX, la pumbosis estaba ampliamente estudiada y las regulaciones consecuentes redujeron hasta 25 veces su ocurrencia desde 1900 hasta 1944. Pese a esto y por irónico que parezca, el mayor pico en la producción de Pb estaba por venir. En 1921 comenzó en Estados Unidos la producción y uso de un tipo de gasolina de alto octanaje que utilizaba [tetraetilo de plomo (TEL)](https://es.wikipedia.org/wiki/Tetraetilo_de_plomo) como antidetonante. Fue recién en la década de 1970, tras ser evidentes los daños que el uso de la gasolina con plomo estaban causando a la salud pública y al medio ambiente, que surgieron las primeras legislaciones en Estados Unidos y algunos países de Europa para reducir la contaminación del aire por Pb. 

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
ds1_icp_ms_pb = pd.read_csv('./data/icp_ms_pb.csv', index_col=0)
ds1_icp_ms_pb.head()
```

```
plt.rcParams["figure.figsize"] = (12, 4.8)
fig, ax = plt.subplots()
ax.plot('Age (Year C.E.)', 'Pb (ng/L)', data=ds1_icp_ms_pb, linewidth=0.5)

x_ticks = [i for i in range(0, 2100, 100)]
ax.set_xticks(x_ticks)

ax.set_xlabel('Year (C.E.)')
ax.set_ylabel('Pb concentration (ng/L)')

plt.show()
```

```
fig, ax = plt.subplots()
ax.plot('Age (Year C.E.)', 'Pb (ng/L)', data=ds1_icp_ms_pb, color='red', linewidth=0.5)

x_ticks = [i for i in range(0, 2100, 100)]
ax.set_xticks(x_ticks)

ax.set_yscale('log')

ax.set_xlabel('Year (C.E.)')
ax.set_ylabel('Pb concentration (ng/L)')

plt.show()
```

```
west_emp_fall = ds1_icp_ms_pb[ds1_icp_ms_pb['Age (Year C.E.)'] <= 476]
middle_ages_condition = np.logical_and((ds1_icp_ms_pb['Age (Year C.E.)'] > 476), (ds1_icp_ms_pb['Age (Year C.E.)'] <= 1453))
middle_ages = ds1_icp_ms_pb[middle_ages_condition]
black_death_condition = np.logical_and((ds1_icp_ms_pb['Age (Year C.E.)'] >= 1347), (ds1_icp_ms_pb['Age (Year C.E.)'] <= 1353))
black_death = ds1_icp_ms_pb[black_death_condition]
renaissance_condition = np.logical_and((ds1_icp_ms_pb['Age (Year C.E.)'] > 1453), (ds1_icp_ms_pb['Age (Year C.E.)'] <= 1750))
renaissance = ds1_icp_ms_pb[renaissance_condition]
ind_rev_condition = np.logical_and((ds1_icp_ms_pb['Age (Year C.E.)'] > 1750), (ds1_icp_ms_pb['Age (Year C.E.)'] <= 1921))
industrial_revolution = ds1_icp_ms_pb[ind_rev_condition]
pb_gas_condition = np.logical_and((ds1_icp_ms_pb['Age (Year C.E.)'] > 1921), (ds1_icp_ms_pb['Age (Year C.E.)'] <= 2000))
pb_gas = ds1_icp_ms_pb[pb_gas_condition]
after_pb_gas =  ds1_icp_ms_pb[ds1_icp_ms_pb['Age (Year C.E.)'] > 2000]

fig, ax = plt.subplots()
ax.plot('Age (Year C.E.)', 'Pb (ng/L)', data=west_emp_fall, color='orange', linewidth=0.5)
ax.plot('Age (Year C.E.)', 'Pb (ng/L)', data=middle_ages, color='green', linewidth=0.5)
ax.plot('Age (Year C.E.)', 'Pb (ng/L)', data=black_death, color='black')
ax.plot('Age (Year C.E.)', 'Pb (ng/L)', data=renaissance, color='blue', linewidth=0.5)
ax.plot('Age (Year C.E.)', 'Pb (ng/L)', data=industrial_revolution, color='pink', linewidth=0.5)
ax.plot('Age (Year C.E.)', 'Pb (ng/L)', data=pb_gas, color='red', linewidth=0.5)
ax.plot('Age (Year C.E.)', 'Pb (ng/L)', data=after_pb_gas, color='black', linewidth=0.5)
ax.set_xticks(x_ticks)
ax.set_yscale('log')
ax.set_xlabel('Year (C.E.)')
ax.set_ylabel('Pb concentration (ng/L)')
plt.show()
```
